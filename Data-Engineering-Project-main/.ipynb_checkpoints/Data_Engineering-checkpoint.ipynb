```python
import os
import shutil
from pathlib import Path
```

# Extracting data for raw layer


```python
downloads_path = str(Path.home() / "Downloads")
notebook_dir = os.getcwd()  
raw_folder = os.path.join(notebook_dir, "raw_folder")  

os.makedirs(raw_folder, exist_ok=True)

files = os.listdir(downloads_path)

for file in files:
    if file.startswith("EmployeeEngagement"):
        shutil.copy(os.path.join(downloads_path, file), raw_folder)

print(f"CSV files have been copied to {raw_folder}")
```

# Starting Preparation Layer

```python
import pandas as pd

# Load the data into dataframes
df_engagements = pd.read_csv('./raw_folder/EmployeeEngagement.engagements.csv')
df_feedbacks = pd.read_csv('./raw_folder/EmployeeEngagement.feedbacks.csv')
df_learningmaterial = pd.read_csv('./raw_folder/EmployeeEngagement.learningmaterials.csv')
df_modules = pd.read_csv('./raw_folder/EmployeeEngagement.modules.csv')
df_quizzes = pd.read_csv('./raw_folder/EmployeeEngagement.quizzes.csv')
df_users = pd.read_csv('./raw_folder/EmployeeEngagement.users.csv')
df_discussion = pd.read_csv('./raw_folder/EmployeeEngagement.discussions.csv')

# Display schema information for each dataframe
print("Schema of df_engagements:")
print(df_engagements.info())  # Displays column names, non-null values, and data types
print("\nSample data of df_engagements:")
print(df_engagements.head())  # Displays first 5 rows for a quick preview

print("\nSchema of df_feedbacks:")
print(df_feedbacks.info())
print("\nSample data of df_feedbacks:")
print(df_feedbacks.head())

print("\nSchema of df_learningmaterial:")
print(df_learningmaterial.info())
print("\nSample data of df_learningmaterial:")
print(df_learningmaterial.head())

print("\nSchema of df_modules:")
print(df_modules.info())
print("\nSample data of df_modules:")
print(df_modules.head())

print("\nSchema of df_quizzes:")
print(df_quizzes.info())
print("\nSample data of df_quizzes:")
print(df_quizzes.head())

print("\nSchema of df_userss:")
print(df_users.info())

print("\nSchema of df_userss:")
print(df_discussion.info())
```

```python
df_learningmaterial
```

```python
# Clean df_engagements
df_engagements['completed'] = df_engagements['learningMaterialCompletion[0].completed'].astype(bool)
df_engagements.drop(columns=[ 
                              'learningMaterialCompletion[0]._id', 
                              'learningMaterialCompletion[0].completed', 
                              '__v'], inplace=True)

# Clean df_feedbacks
df_feedbacks['createdAt'] = pd.to_datetime(df_feedbacks['createdAt'])
df_feedbacks.drop(columns=['__v'], inplace=True)

# Clean df_learningmaterial
df_learningmaterial['createdAt'] = pd.to_datetime(df_learningmaterial['createdAt'])
df_learningmaterial.drop(columns=['__v'], inplace=True)

# Clean df_modules
df_modules.drop(columns=['__v'], inplace=True)

df_quizzes.drop(columns=['__v'], inplace=True)

df_users.drop(columns=['__v'], inplace=True)
```

```python
df_users.rename(columns={
    '_id': 'userId',
}, inplace=True)

df_learningmaterial.rename(columns={
    '_id': 'learningMaterialId',
}, inplace=True)

df_engagements.rename(columns={
    'learningMaterialCompletion[0].learningMaterialId': 'learningMaterialId',
}, inplace=True)
```

```python
df_users
```

# Dimension Tables
# 1. Users Dimension Table
# 2. Learning Material Dimension Table
# 3. Modules Dimension Table
# 4. Quizzes Dimension Table

```python
# Users Dimension

dim_users = df_users[['userId', 'firstName', 'lastName','department','role']].drop_duplicates()

# Learning Material Dimension
dim_learning_material = df_learningmaterial[['learningMaterialId', 'title', 'description', 'duration']].drop_duplicates()

# Modules Dimension
dim_modules = df_modules[['learningMaterialId', 'title', 'description']].drop_duplicates()

# Quizzes Dimension
dim_quizzes = df_quizzes[['learningMaterialId', 'title']].drop_duplicates()
```

```python
df_engagements
```

```python
# Creating the Engagement Fact Table



fact_engagements = df_engagements[['userId', 'completed', 'quizScore','learningMaterialId']].copy()  # Use copy() to avoid the warning
fact_engagements['engagement_id'] = df_engagements['_id'].values  # Use .values to ensure you're working with the underlying array

# Feedback Fact Table
fact_feedbacks = df_feedbacks[['userId', 'learningMaterialId', 'rating']].copy()  # Use copy() here as well

```

```python
fact_engagements
```

```python
dataframes = {
    "fact_engagements.csv":fact_engagements,
    "dim_quizzes.csv": dim_quizzes,
    "fact_feedbacks.csv": fact_feedbacks,
    "dim_modules.csv" : dim_modules,
    "dim_learningmaterials.csv" :dim_learning_material,
    "dim_users.csv" :dim_users
}


prep = "prep"

for name, df in dataframes.items():
    print(f"Schema for {name}:")
    print(df.dtypes)
    print("\n")
if not os.path.exists(prep):
    os.makedirs(prep)

for filename, df in dataframes.items():
    file_path = os.path.join(prep, filename)
    df.to_csv(file_path, index=False) 
print(f"All DataFrames have been saved in the '{prep}' folder.")
```

# User Engagement Summary

```python
user_engagement_summary = fact_engagements.groupby('userId').agg(
    total_engagements=('engagement_id', 'count'),
    avg_quiz_score=('quizScore', 'mean'),
    completed_quizzes=('completed', 'sum')
).reset_index()

user_engagement_summary
```

# Feedback Summary

```python

feedback_summary = fact_feedbacks.groupby('learningMaterialId').agg(
    average_rating=('rating', 'mean'),
    total_feedbacks=('rating', 'count')
).reset_index()


feedback_summary = feedback_summary.merge(
    dim_learning_material[['learningMaterialId', 'title']], on='learningMaterialId', how='left'
)

feedback_summary
```

#  User Feedback Summary

```python
user_feedback_summary = fact_feedbacks.groupby('userId').agg(
    total_feedbacks=('rating', 'count'),
    average_rating=('rating', 'mean')
).reset_index()

# Merge with dimension table for user details
user_feedback_summary = user_feedback_summary.merge(
    dim_users, on='userId', how='left'
)
user_feedback_summary
```

# User Performance Report

```python
# Merge feedbacks with engagements
user_performance_report = df_feedbacks.merge(
    df_engagements, on='userId', how='left'
).merge(
    dim_users[['userId', 'firstName', 'lastName']], on='userId', how='left'
).groupby('userId').agg(
    first_name=('firstName', 'first'),
    last_name=('lastName', 'first'),
    average_quiz_score=('quizScore', 'mean'),
    average_feedback_rating=('rating', 'mean')
).reset_index()


user_performance_report
```

```python
# # Step 1: Merge fact_engagements with dim_users to include department
# fact_engagements_with_dept = fact_engagements.merge(
#     dim_users[['userId', 'firstName', 'lastName', 'department']], 
#     on='userId', 
#     how='left'
# )

# fact_engagements_with_dept['engagement_score'] = fact_engagements_with_dept['quizScore'] + fact_engagements_with_dept['completed'].astype(int)

# engagement_summary = fact_engagements_with_dept.groupby(['userId', 'firstName', 'lastName', 'department']).agg(
#     total_engagement_score=('engagement_score', 'sum'),
#     total_engagements=('completed', 'count')
# ).reset_index()

# top_employees = engagement_summary.nlargest(5, 'total_engagement_score')
# lowest_employees = engagement_summary.nsmallest(5, 'total_engagement_score')
# Step 1: Merge fact_engagements with dim_users to include department

fact_engagements
fact_engagements_with_dept = fact_engagements[['userId', 'quizScore', 'completed', 'learningMaterialId']].merge(
    dim_users[['userId', 'firstName', 'lastName', 'department']], 
    on='userId', 
    how='left'
)


# Step 2: Calculate Engagement Score
fact_engagements_with_dept['engagement_score'] = fact_engagements_with_dept['quizScore'] + fact_engagements_with_dept['completed'].astype(int)

# Step 3: Merge with dim_learningmaterials to include learning material title
fact_engagements_with_dept = fact_engagements_with_dept.merge(
    dim_learning_material[['learningMaterialId', 'title']], 
    on='learningMaterialId', 
    how='left'
)

# Step 4: Aggregate Scores by User, including department, quiz scores, completed, and learning material title
engagement_summary = fact_engagements_with_dept.groupby(
    ['userId', 'firstName', 'lastName', 'department', 'title']
).agg(
    total_engagement_score=('engagement_score', 'sum'),
    total_quiz_score=('quizScore', 'sum'),
    total_completed=('completed', 'sum')
).reset_index()

# Step 5: Identify Top and Lowest Employees
top_employees = engagement_summary.nlargest(5, 'total_engagement_score')
lowest_employees = engagement_summary.nsmallest(5, 'total_engagement_score')

# Print Results
print("Top Employees based on Engagement:")
print(top_employees)

print("\nLowest Employees based on Engagement:")
print(lowest_employees)
```
