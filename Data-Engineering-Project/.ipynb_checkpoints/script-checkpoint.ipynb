```python
import os
import shutil
from pathlib import Path
```

# Extracting data for raw layer


```python
downloads_path = str(Path.home() / "Downloads")
notebook_dir = os.getcwd()  
raw_folder = os.path.join(notebook_dir, "raw_folder")  

os.makedirs(raw_folder, exist_ok=True)

files = os.listdir(downloads_path)

for file in files:
    if file.startswith("majorProject"):
        shutil.copy(os.path.join(downloads_path, file), raw_folder)

print(f"CSV files have been copied to {raw_folder}")
```

# Starting Preparation Layer

```python
import pandas as pd

df_engagements = pd.read_csv('./raw_folder/majorProject.engagements.csv')
df_feedbacks = pd.read_csv('./raw_folder/majorProject.feedbacks.csv')
df_learningmaterial = pd.read_csv('./raw_folder/majorProject.learningmaterials.csv')
df_modules = pd.read_csv('./raw_folder/majorProject.modules.csv')
df_quizzes = pd.read_csv('./raw_folder/majorProject.quizzes.csv')
df_userss = pd.read_csv('./raw_folder/majorProject.userss.csv')
```

# Mostly data is cleaned except engagements table

```python
selected_engagements_columns = ['_id', 'userId','learningMaterialCompletion[0].learningMaterialId','quizScore']
df_selected_engagements = df_engagements[selected_engagements_columns]
```

```python
df_selected_engagements = df_selected_engagements.rename(columns={
    'learningMaterialCompletion[0].learningMaterialId': 'learningMaterialId'  
})
df_selected_engagements
```

```python
selected_quizzes_columns = ['_id','title','learningMaterialId']
df_selected_quizzes =  df_quizzes[selected_quizzes_columns]
```

```python
dataframes = {
    "prep_engagements.csv":df_selected_engagements ,
    "prep_quizzes.csv": df_selected_quizzes,
    "prep_feedbacks.csv": df_feedbacks,
    "prep_modules.csv" : df_modules,
    "prep_learningmaterials.csv" :df_learningmaterial,
    "prep_users.csv" :df_userss
}

prep_folder = "prep_folders"

if not os.path.exists(prep_folder):
    os.makedirs(prep_folder)

for filename, df in dataframes.items():
    file_path = os.path.join(prep_folder, filename)
    df.to_csv(file_path, index=False) 
print(f"All DataFrames have been saved in the '{prep_folder}' folder.")
```

# Reporting Tables

```python
def get_top_employees(df_selected_engagements, df_userss):
 
    top_employees = df_selected_engagements.groupby('userId').agg(totalScore=('quizScore', 'sum')).reset_index()
    
    top_employees = top_employees.sort_values(by='totalScore', ascending=False).head(10)
    
    top_employees = top_employees.merge(
        df_userss[['firstName', 'lastName', 'department', '_id']], 
        left_on='userId', 
        right_on='_id', 
        suffixes=('', '_user')
    )
    
    top_employees = top_employees[['userId', 'firstName', 'lastName', 'department', 'totalScore']]
    
    return top_employees

def get_department_scores(df_selected_engagements, df_users):
    merged_df = df_selected_engagements.merge(df_userss, left_on='userId', right_on='_id', suffixes=('', '_user'))
    
    department_scores = merged_df.groupby('department').agg(
        averageScore=('quizScore', 'mean'),
        totalScore=('quizScore', 'sum'),
        count=('quizScore', 'size')
    ).reset_index()
    
    return department_scores
    


    
def get_top_feedback_learning_material(df_feedbacks, df_learningmaterial):
    top_material_feedbacks = df_feedbacks.groupby('learningMaterialId').agg(
        totalFeedbacks=('rating', 'count'),
        averageRating=('rating', 'mean')
    ).reset_index()
    
    top_material_feedbacks = top_material_feedbacks.sort_values(by='totalFeedbacks', ascending=False).head(1)
    
    top_material_feedbacks = top_material_feedbacks.merge(
        df_learningmaterial[['title', '_id']],
        left_on='learningMaterialId',
        right_on='_id',
        suffixes=('', '_material')
    )
    
    top_material_feedbacks = top_material_feedbacks[['learningMaterialId', 'title', 'totalFeedbacks', 'averageRating']]
    
    return top_material_feedbacks


top_employees = get_top_employees(df_selected_engagements,df_userss)
department_scores = get_department_scores(df_selected_engagements, df_userss)
top_learning_material = get_top_feedback_learning_material(df_feedbacks, df_learningmaterial)

print("Top Employees:\n", top_employees)
print("\nDepartment Scores:\n", department_scores)
print("\nTop Learning Material:\n", top_learning_material)


Reprtingdataframes = {
    "top_employees.csv":top_employees ,
    "department_scores.csv": department_scores,
    "top_learning_material.csv": top_learning_material,
}

Reporting = "Reporting"

if not os.path.exists(Reporting):
    os.makedirs(Reporting)

for filename, df in Reprtingdataframes.items():
    file_path = os.path.join(Reporting , filename)
    df.to_csv(file_path, index=False) 
print(f"All DataFrames have been saved in the '{Reporting }' folder.")

```
